```python
import matplotlib.pyplot as plt
import seaborn as sns

# Clean column names
df.columns = df.columns.str.strip()

# Group and sort
industry_emission = df.groupby('Industry Name')['Supply Chain Emission Factors with Margins'].sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=industry_emission.values, y=industry_emission.index)
plt.title("Top 10 Industries by Supply Chain Emission (with Margins)")
plt.xlabel("Total Emission Factor (with Margins)")
plt.ylabel("Industry Name")
plt.tight_layout()
plt.show()
```
